# Check for duplicates
Sometimes scrapers put the same record more than once into a csv file.<br>
This script is not fully tested, because there were no duplicates since I finished it.

```python
import pandas as pd
```

```python
cantons = ['AG','AI','AR','BE','BL','BS','FR','GE','GL','GR','JU','LU','NE',
           'NW','OW','SG','SH','SO','SZ','TG','TI','UR','VD','VS','ZG','ZH']

url_head = 'https://raw.githubusercontent.com/openZH/covid_19/master/fallzahlen_kanton_total_csv/COVID19_Fallzahlen_Kanton_'
url_tail = '_total.csv'

# create empty list for DataFrames
dataframes = []

for c in cantons:
    url = url_head + c + url_tail
    df_canton = pd.read_csv(url)
    dataframes.append(df_canton)
    
df_CH = pd.concat(dataframes, sort=False) 
df_CH.rename(columns={'abbreviation_canton_and_fl': 'canton'},inplace=True)
df_CH.head()
```

```python
# test for duplicate entries 
dupl = df_CH[df_CH.duplicated(['date','canton'], keep=False)]

if len(dupl.index) > 0:
    df_CH.drop_duplicates(['Date','canton'], keep='last',inplace=True)
    dupl.head()
else:
    print("there are no duplicates")
```
